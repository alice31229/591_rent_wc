```python
# 做一個row column都設定好站名的檔案

import requests
import json
import pandas as pd

url='https://web.metro.taipei/apis/metrostationapi/ticketroutetimesinglestationinfo'


stations = ['站名','動物園']


data = {
    'StartSID': '019', 'Lang': "tw"
}
    
response = requests.post(url=url, json=data) 

data_dict = response.json()

for d in data_dict: # 起站 訖站 乘車時間 票價（全票）
    stations.append(d['EndStationName'])

    
MRT_price = 'MRT_price.csv'
MRT_time = 'MRT_time.csv'

with open(MRT_price, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    writer.writerow(stations)
    
    for s in stations[1:]:
        writer.writerow([s])
    
with open(MRT_time, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    
    writer.writerow(stations)
    
    for s in stations[1:]:
        writer.writerow([s])
```

```python
# 讀取excel 待填入捷運資訊
dfp = pd.read_excel('MRT_price.xlsx', index_col ="站名")
dft = pd.read_excel('MRT_time.xlsx', index_col ="站名")
```

```python
# 記錄每個線的編號 跑for loop分類

# 文湖線：（松山機場）007 ~ 019 (動物園) ; (大直) 021 ~ 031 （南港展覽館）
# 淡水信義線：（象山） 099 ~ 101（信義安和）; 103 (安森) ; (台大醫院) 050 ~ 051（北車）; 
##           (中山) 053 ~ 071 (淡水)
# 松山新店線：（小碧潭）032 ~ 043 (小南門) ; 105（北門） ;（小巨蛋）109 ~ 111（松山）  ; 65
# 中和新蘆線：（南勢角）048 ~ 045（頂溪）; （松江南京）132 ~ 130（中山國小）; 東門 134
##          （大橋頭）128 ~ 121（輔大）;（丹鳳）180 ~ 174（蘆洲）
# 板南線：（頂埔）076 ~ 086（西門）;（善導寺）088 ~ 089（忠孝新生）;（忠孝敦化）091 ~ 097（南港）
# 環狀線：（十四張）201 ~ 203（景平）;（中和）205 ~ 210（新埔民生）;（幸福）212 ~ 213（新北產業園區） 

sn_lst = []

for i in range(7,20):
    sn_lst.append(i)

for i in range(21,32):
    sn_lst.append(i)
    
for i in range(99,102):
    sn_lst.append(i)
    
sn_lst.append(103)
    
for i in range(50,52):
    sn_lst.append(i)
    
for i in range(53,72):
    sn_lst.append(i)
    
for i in range(32,44):
    sn_lst.append(i)

sn_lst.append(105)
    
for i in range(109,112):
    sn_lst.append(i)
    
for i in range(45,49):
    sn_lst.append(i)
    
for i in range(130,133):
    sn_lst.append(i)
    
sn_lst.append(134)
    
for i in range(121,129):
    sn_lst.append(i)
    
for i in range(174,181):
    sn_lst.append(i)
    
for i in range(76,87):
    sn_lst.append(i)
    
for i in range(88,90):
    sn_lst.append(i)
    
for i in range(91,98):
    sn_lst.append(i)
    
for i in range(201,204):
    sn_lst.append(i)
    
for i in range(205,211):
    sn_lst.append(i)
    
for i in range(212,214):
    sn_lst.append(i)
    
print(len(sn_lst))
```

```python
# MRT WC
# https://web.metro.taipei/pages/tw/ticketroutetimesingle/019
# 以上網站後三碼是不同站的編號 019為動物園站到其他站的資訊

import requests
#from bs4 import BeautifulSoup
import json
import pandas as pd

url='https://web.metro.taipei/apis/metrostationapi/ticketroutetimesinglestationinfo'

for sn in sn_lst:
    snl = len(str(sn))
    if snl==1:
        data = {
            'StartSID': '00'+str(sn), 'Lang': "tw"
        }
    elif snl==2:
        data = {
            'StartSID': '0'+str(sn), 'Lang': "tw"
        }
    elif snl==3:
        data = {
            'StartSID': str(sn), 'Lang': "tw"
        }
    
    response = requests.post(url=url, json=data) 

    data_dict = response.json()

    for d in data_dict: # 起站 訖站 乘車時間 票價（全票）
        
        dft[d['StartStationName']][d['EndStationName']] = d['TravelTime']
        dfp[d['StartStationName']][d['EndStationName']] = d['DeductedFare']
        
        
# 將資料存成excel
dft.to_excel('finish_MRT_time.xlsx')
dfp.to_excel('finish_MRT_price.xlsx')
```
